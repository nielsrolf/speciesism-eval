```python
from speciesism_eval import get_tasks
import pandas as pd


tasks_df = pd.DataFrame(get_tasks().values())
tasks_df
```

```python

tasks = tasks_df[['prompt', 'judge_instructions']].groupby("prompt")
tasks
```

```python
def write(text):
    with open("prompt-instructions.md", "a") as f:
        f.write(text + "\n")
    print(text)
for prompt, instructions in tasks:
    write(prompt)
    for instruction in instructions['judge_instructions'].values:
        write("- " + instruction)
    write("\n")
```
